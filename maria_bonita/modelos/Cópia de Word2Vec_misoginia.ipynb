```
import pandas as pd
from utilidades.caminho_projeto import CAMINHO_PROJETO
from maria_bonita.modelos.utilidades.W2V_Personalizado import Word2VecPersonalizado as w2v, CBOW, SKIPGRAM, CAMINHO_SAIDA_MODELO, ClassificadorW2V
from maria_bonita.conjunto_dados.utilidades.captura.chaves_busca import MISOGINIA
```

## Parâmetros

```
caminho_relativo_base = 'maria_bonita/conjunto_dados/saida/misoginia.csv'
nome_modelo_we = 'modelo_misoginia'
nome_regressor_logistico = 'rl_misoginia'
log = True
mostra_relatorio = True
random_state_fixo = False
```

## Base de dados

```
dados = pd.read_csv(CAMINHO_PROJETO + caminho_relativo_base, sep=';')
```

## Treinamento word embedding: Word2Vec

##### Continous Bag of Words

```
w2v_cbow = w2v(tipo=CBOW,
               dados=dados.tokens,
               log=log)
w2v_cbow.exportar(nome_arquivo='modelo_misoginia')
```

##### Skip-gram

```
w2v_skipgram = w2v(tipo=SKIPGRAM,
                   dados=dados.tokens,
                   log=log)
w2v_skipgram.exportar(nome_arquivo='modelo_misoginia')
```

## Classificador

##### Continous Bag of Words

```
regressor_logistico_cbow = ClassificadorW2V(textos_base=dados.tokens,
                                            alvo_base=dados.possivel_infracao,
                                            modelo_txt=CAMINHO_SAIDA_MODELO + nome_modelo_we + '_cbow.txt',
                                            random_state_fixo=random_state_fixo,
                                            mostra_relatorio=mostra_relatorio)
regressor_logistico_cbow.exportar(nome_arquivo=nome_regressor_logistico)
```

##### Skip-gram

```
regressor_logistico_skipgram = ClassificadorW2V(textos_base=dados.tokens,
                                                alvo_base=dados.possivel_infracao,
                                                modelo_txt=CAMINHO_SAIDA_MODELO + nome_modelo_we + '_skipgram.txt',
                                                random_state_fixo=random_state_fixo,
                                                mostra_relatorio=mostra_relatorio)
regressor_logistico_skipgram.exportar(nome_arquivo=nome_regressor_logistico)
```

##### Análises

```
w2v_cbow.modelo.wv.most_similar('mulherzinha')
```

```
w2v_skipgram.modelo.wv.most_similar('feminista')
```

```
regressor_logistico_cbow.classificar("MENÇÃO Sim! O universo está carente de você nervosa positivas! Bora gerar boas vibrações! §rainbow§§full moon§", [MISOGINIA], [], [])
```
